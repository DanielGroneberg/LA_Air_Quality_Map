## Converting .csv points to a raster file

I will attempt to implement this code:

https://github.com/neerajn07/Drone-Geophysics/blob/main/csv2tiff.py

```python
import pandas as pd
import numpy as np
from scipy.interpolate import griddata
import rasterio
from rasterio.transform import from_origin
import os

def csv_to_geotiff(
    csv_path,
    x_col='X',
    y_col='Y',
    z_col='Z',
    output_tif='output.tif',
    interp_method='linear',
    grid_res=12000,
    epsg_code=32723
):
    # Check file existence
    if not os.path.exists(csv_path):
        raise FileNotFoundError(f"CSV file not found: {csv_path}")

    # Read CSV
    try:
        df = pd.read_csv(csv_path)
    except Exception as e:
        raise RuntimeError(f"Error reading CSV: {e}")
    
    # Check column presence
    for col in [x_col, y_col, z_col]:
        if col not in df.columns:
            raise ValueError(f"Column '{col}' not found in CSV.")

    # Drop NaN rows
    df = df[[x_col, y_col, z_col]].dropna()
    
    # Extract coordinates
    x = df[x_col].values
    y = df[y_col].values
    z = df[z_col].values

    if len(x) < 3:
        raise ValueError("Not enough points to perform interpolation.")

    # Generate interpolation grid
    xi = np.arange(x.min(), x.max(), grid_res)
    yi = np.arange(y.min(), y.max(), grid_res)
    xi, yi = np.meshgrid(xi, yi)

    # Interpolate
    print(f"Interpolating using method: {interp_method}")
    try:
        zi = griddata((x, y), z, (xi, yi), method=interp_method)
    except Exception as e:
        raise RuntimeError(f"Interpolation failed: {e}")

    # Handle NaNs if any
    nan_count = np.isnan(zi).sum()
    if nan_count > 0:
        print(f"Warning: {nan_count} NaN values in the interpolated grid. Filling with nearest.")
        zi = griddata((x, y), z, (xi, yi), method='nearest')

    # Define affine transform
    transform = from_origin(x.min(), y.max(), grid_res, grid_res)

    # Write to GeoTIFF
    try:
        with rasterio.open(
            output_tif,
            'w',
            driver='GTiff',
            height=zi.shape[0],
            width=zi.shape[1],
            count=1,
            dtype='float32',
            crs=f'EPSG:{epsg_code}',
            transform=transform
        ) as dst:
            dst.write(zi.astype('float32'), 1)
        print(f"GeoTIFF saved at: {output_tif}")
    except Exception as e:
        raise RuntimeError(f"Failed to write GeoTIFF: {e}")
```

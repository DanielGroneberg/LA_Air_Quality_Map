I'll try pulling in from this station on July 3rd, 2021 - simply because that's within the date range

```python
url = "https://openaq-data-archive.s3.amazonaws.com/records/csv.gz/locationid=847/year=2021/month=07/location-847-20210703.csv.gz"
```

```python
def download_file(url, save_path):
    response = requests.get(url, stream=True)
    
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f'{url} downloaded')
    else:
        print(f'{url} failed')
```

```python
download_file(url,"Data/Inputs/station_847.csv")
```
